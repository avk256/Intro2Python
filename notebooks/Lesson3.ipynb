# Умовні оператори в Python




## **`if`, `else`, `elif`**

Python використовує ключове слово `if` для перевірки логічних умов. Якщо умова істинна (`True`), виконується відповідний блок коду.


```python
x = 10
print(x > 0)
if x > 0:
    print("Додатнє число")
else:
    print("Нуль або від’ємне")
```


## **`if-elif-else`**: перевірка кількох варіантів

* **`if`**: перевіряється перша умова.
* **`elif`**: якщо попередні умови не спрацювали, перевіряється ця.
* **`else`**: якщо жодна умова не виконалась — спрацьовує блок `else`.




```python
grade = 20


if grade >= 90:
    print("A")
elif grade >= 75:
    print("B")
elif grade >= 60:
    print("C")
else:
    print("F")
```

## `is` vs `==`

* `==` перевіряє **рівність значень**.
* `is` перевіряє **ідентичність об’єктів** (чи це одне й те ж саме в пам’яті).




```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(id(a))
print(id(b))
print(id(c))

print((a == b) and (a is b))
print(a is c)
print(a is b)
```

## `match-case` (Python 3.10+)

Аналог `switch` з інших мов. Дозволяє розгалуження за значенням або структурою.




```python
command = "stop"

match command:
    case "start":
        print("Запуск")
    case "stop":
        print("Зупинка")
    case _:
        print("Невідома команда")
```

Можна також додавати **guard-умови**:



```python
match (x := 6):
    case n if n % 2 == 0:
        print("Парне")
    case _:
        print("Непарне")
```

## `enumerate()` — доступ до індексу



```python

colors = ["red", "green", "blue"]

print(list(enumerate(colors)))
idx_dict = dict(list(enumerate(colors)))

print(idx_dict)

for idx, color in enumerate(colors):
    print(f"{idx}: {color}")
```

## `zip()` — одночасний перебір кількох послідовностей




```python
names = ["Anna", "Ben", "Carla"]
ages = [25, 30, 28]

print(list(zip(names, ages)))

my_dict = dict(list(zip(names, ages)))

print(my_dict)

print(my_dict.keys())

for name, age in zip(names, ages):
    print(f"{name} is {age} years old")
```

# Обробка винятків (`try-except`)

Коли виникає помилка (виняток), Python може припинити виконання. Щоб цього уникнути, використовують `try-except`.





## Базовий приклад


```python
try:
    number = int(input("Введіть число: "))
    print("Ви ввели:", number)
except ValueError:
    print("Це не число!")
```

###ZeroDivisionError

Виникає при спробі поділити на нуль.

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Не можна ділити на нуль")
```

### ValueError

Невідповідний тип значення (наприклад, спроба перетворити рядок, що не є числом, у int).

```python
try:
    num = int("abc")
except ValueError:
    print("Неможливо перетворити рядок у число")
```

### TypeError

Невідповідний тип операндів або аргументів (наприклад, додавання рядка до числа).

```python
try:
    result = "5" + 3
except TypeError:
    print("Не можна додавати рядок до числа")
```

### IndexError

Індекс за межами списку.

```python
try:
    lst = [1, 2, 3]
    print(lst[5])
except IndexError:
    print("Індекс за межами списку")
```

### KeyError

Невірний ключ при доступі до словника.

```python
try:
    d = {"ім'я": "Олег"}
    print(d["вік"])
except KeyError:
    print("Ключ не знайдено у словнику")
```

###NameError

Звернення до змінної, яка ще не визначена.

```python
try:
    print(невизначена_змінна)
except NameError:
    print("Змінну не знайдено")
```

###AttributeError

Спроба звернутися до неіснуючого атрибута об'єкта.

```python
try:
    x = 5
    x.append(10)
except AttributeError:
    print("Цей об'єкт не має методу append()")
```

## Перехоплення кількох типів помилок:




```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Ділення на нуль")
except Exception as e:
    print("Інша помилка:", e)
```

## `else` і `finally`:



```python
try:
    x = int(input("Введіть число: "))
    result = 10 / x
except ZeroDivisionError:
    print("Помилка: ділення на нуль")
except ValueError:
    print("Помилка: введено не число")
else:
    print("Результат:", result)
finally:
    print("Цей блок виконується завжди")
```

try — спроба виконати код.

except — перехоплення помилок.

else — виконується, якщо помилок не було.

finally — виконується завжди, навіть якщо стався виняток.
