#Множини

set — змінювана невпорядкована колекція унікальних хешованих елементів.

Перевірка in, додавання та видалення зазвичай O(1).

Літерал: {1, 2, 3}. Порожня множина: set() (бо {} — це порожній словник).

## Алгебра множин

![picture](https://drive.google.com/uc?id=1lGCwwOoCdHEpff0qhixQ1-a_JD_Niugc)



```python
from pprint import pprint

# Базові множини
A = {1, 2, 3, 4, 5}
B = {3, 4, 5, 6}
C = {5, 6, 7}

print(" Початкові множини ")
print("A =", A)
print("B =", B)
print("C =", C)

# ---------------------------
# ОПЕРАЦІЇ, ЩО ПОВЕРТАЮТЬ НОВУ МНОЖИНУ
# ---------------------------

print("\nОПЕРАЦІЇ (новий результат) ")
U = A.union(B)                    # A ∪ B
print("A.union(B)                 ->", U)

U3 = A.union(B, C)                # A ∪ B ∪ C (можна передати кілька ітерабельних)
print("A.union(B, C)              ->", U3)

I = A.intersection(B)             # A ∩ B
print("A.intersection(B)          ->", I)

I3 = A.intersection(B, C)         # A ∩ B ∩ C
print("A.intersection(B, C)       ->", I3)

D = A.difference(B)               # A \ B
print("A.difference(B)            ->", D)

D3 = A.difference(B, C)           # A \ (B ∪ C) — елементи A, що не в B і не в C
print("A.difference(B, C)         ->", D3)

S = A.symmetric_difference(B)     # (A \ B) ∪ (B \ A)
print("A.symmetric_difference(B)  ->", S)
```

```python

A = {1, 2, 3, 4}
B = {3, 4, 5}
print("A:", A, "B:", B)
print("A | B  (об'єднання):       ", A | B)
print("A & B  (перетин):          ", A & B)
print("A - B  (різниця):          ", A - B)
print("A ^ B (симетрична різниця):", A ^ B)

# версії "на місці"
C = A.copy()
C.intersection_update(B)       # C &= B
print("intersection_update ->", C)
D = A.copy()
D.difference_update(B)         # D -= B
print("difference_update ->", D)
```

```python
print("\n FROZENSET (незмінювана множина) ")
FA = frozenset(A)
FB = frozenset(B)
print("FA.union(FB)               ->", FA.union(FB))         # новий frozenset
print("FA.intersection(FB)        ->", FA.intersection(FB))
print("FA.difference(FB)          ->", FA.difference(FB))
print("FA.isdisjoint(FB)          ->", FA.isdisjoint(FB))



```

#Кортежі

Кортеж — впорядкована незмінювана послідовність.

Добре підходить для “записів” фіксованої структури (наприклад, ('Kyiv', 3_000_000)), для ключів у словниках та елементів множин.

Методи самого tuple: тільки .count(x) і .index(x). Решта — через оператори та вбудовані функції (len, sum, min, max, sorted, зрізи тощо).

Літерал: (1, 2, 3) або просто 1, 2, 3. Один елемент: (42,) — кома обовʼязкова. Порожній: ().

## Створення

```python

t1 = (1, 2, 3)
t2 = 4, 5, 6
single = (42,)
empty = ()
nested = (1, (2, 3), ["a", "b"])

print("t1:", t1)
print("t2:", t2)
print("single:", single, "len:", len(single))
print("empty:", empty)
print("nested:", nested)

pprint(dir(t1))
```

# Зрізи


## Теорія

**Синтаксис** `seq[start:stop:step]`

* `start` — індекс першого елемента (входить у зріз).
* `stop` — **межа зверху** (не входить у зріз).
* `step` — крок (може бути від’ємним).
  Будь-який з параметрів можна опустити: `[:]`, `[:stop]`, `[start:]`, `[::step]`.

**Еквівалентність із `range`:** елементи зрізу — це `seq[i]` для `i` з `range(start, stop, step)`.

**Індекси**

* Дозволені **негативні** індекси (‐1 — останній елемент).
* Індекси, що виходять за межі, **обрізаються** (не `IndexError` для зрізів).
* `step=0` заборонено (`ValueError`).

**Тип результату** такий самий “послідовнісний” тип:

* list → list, str → str, tuple → tuple, range → range.

**(Не)змінність і присвоєння зрізу:**

* Рядки й кортежі — **immutable**: їхні зрізи — нові об’єкти; присвоєння `s[i:j]=...` неможливе.
* Списки — **mutable**:

`lst[i:j] = iterable` замінює ділянку;

`lst[i:j] = []` видаляє;

`lst[i:i] = ...` вставляє.

  * Якщо `step != 1`, довжина правої частини має збігатися з кількістю позицій зрізу.

**Приклади**

* `seq[:]` → копія всього.
* `seq[:n]` → від початку до `n` (не включно).
* `seq[-n:]` → останні `n`.
* `seq[::2]` → кожен 2-й елемент.
* `seq[::-1]` → реверс.


* **Рядки/кортежі**

зрізи завжди створюють **новий** об’єкт (бо вони immutable).
* **Списки**

зріз — **нова** часткова копія; елементи всередині — ті самі об’єкти (мілка копія).
* `lst[::-1]` — зручно для отримання перевернутої **копії**; для ін-плейс реверсу — `lst.reverse()`.


```python
lst = [0, 1, 2, 3, 4, 5, 6]
s  = "Hello, world!"
t  = (10, 20, 30, 40, 50)

print("lst:", lst)
print("s  :", s)
print("t  :", t)
```

```python
# start:stop (stop виключено)
print("\nЗрізи без кроку:")
print("lst[1:4]    ->", lst[1:4])
print("s[:5]       ->", s[:5])
print("t[2:]       ->", t[2:])
print("lst[-3:]    ->", lst[-3:])
print("lst[:100]   ->", lst[:100])
```

```python
# Копія всього (shallow copy для списку)
print("\nКопія послідовності:")
print("lst[:]      ->", lst[:])

print("\nКрок (step):")
print("lst[::2]    ->", lst[::2])      # кожен 2-й
print("lst[1::2]   ->", lst[1::2])     # з індексу 1 через один
print("s[::-1]     ->", s[::-1])       # реверс (НОВА строка)
print("t[::-1]     ->", t[::-1])       # реверс (НОВИЙ кортеж)
print("lst[5:1:-1] ->", lst[5:1:-1])   # крок від’ємний: [5,4,3,2]

try:
    _ = lst[::0]
except ValueError as e:
    print("lst[::0] -> ValueError:", e)
```

```python
# Присвоєння зрізу - лише для списків
a = [10, 20, 30, 40, 50]
print("a старт:", a)

# Заміна піддіапазону (довжина може змінитися)
a[1:4] = [200, 300]                     # 20,30,40 -> 200,300
print("a[1:4]=[200,300] ->", a)

# Вставка без видалення: зріз нульової довжини
a[2:2] = [999, 1000]                    # вставка перед індексом 2
print("a[2:2]=[999,1000] ->", a)

# Видалення піддіапазону
a[1:4] = []
print("a[1:4]=[] ->", a)

# Присвоєння з кроком: довжини повинні збігатися
b = [0, 1, 2, 3, 4, 5, 6]
print("\nb старт:", b)
b[::2] = [100, 200, 300, 400]           # позиції 0,2,4,6
print("b[::2]=[100,200,300,400] ->", b)
try:
    b[::2] = [1, 2]                     # замало елементів
except ValueError as e:
    print("помилка довжин для step!=1:", e)

# Видалення з кроком
c = [0, 1, 2, 3, 4, 5, 6, 7]
del c[1::2]                              # прибрати всі елементи на непарних індексах
print("del c[1::2] ->", c)


```

```python
arr = list(range(10))
print("arr:", arr)

# Реверс (копія) vs ін-плейс:
rev_copy = arr[::-1]
arr_copy = arr[:]
arr_copy.reverse()                       # на місці
print("rev_copy:", rev_copy, "in-place:", arr_copy)

# Останні N:
print("останні 3:", arr[-3:])

# Кожен n-й елемент:
print("кожен 3-й:", arr[::3])

# Видалити перший і останній елемент:
mid = arr[1:-1]
print("без країв:", mid)


```
