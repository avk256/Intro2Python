##Десяткова → двійкова / шістнадцяткова

```python
print(bin(18))          # '0b11111111'         ← двійковий з префіксом 0b
print(hex(255))          # '0xff'               ← шістнадцятковий з префіксом 0x
print(format(255, 'b'))  # '11111111'           ← без префікса
print(format(255, '08b'))# '11111111' з паддінгом до 8 розрядів
print(format(255, 'x'))  # 'ff'                 ← нижній регістр
print(format(255, 'X'))  # 'FF'                 ← верхній регістр
```

##Двійкова / шістнадцяткова → десяткова

```python
print(int('11111111', 2))    # 255
print(int('ff', 16))         # 255
print(int('0b11111111', 0))  # 255  ← base=0: авто-визначення за префіксом
print(int('0xff', 0))        # 255
```

#Форматований вивід через f-рядки

```python
name = "Ada"
score = 93.4567
n = 255

print(f"Вітаю, {name}!")
print(f"Округлення до 2 знаків: {score:.2f}")
print(f"Шістнадцяткове: {n:#x}  Бінарне: {n:08b}")
```

#Розміри типів

```python
import sys
small = 10
big = 10**100
print(sys.getsizeof(small), "байт для small int")
print(sys.getsizeof(big),   "байт для big int")

f = 3.14
z = 1+2j
print(sys.getsizeof(f), "байт для float")
print(sys.getsizeof(z), "байт для complex")
```
